```python
import os
import cv2
import json
import matplotlib.pyplot as plt

from ultralytics import YOLO
```

```python
#Load a model
yolo_yaml_path = '/content/human_detection_dataset/data.yaml'
model = YOLO('yolov8s.yaml').load('yolov8s.pt')

epochs = 15
imgsz = 640

results = model.train(
    data = yolo_yaml_path,
    epochs = epochs,
    imgsz = imgsz,
    project = 'models',
    name = 'yolov8/detect/train',
)
```

```python
model_path = '/content/models/yolov8/detect/train3/weights/best.pt'
model = YOLO(model_path)

metrics = model.val(
    project= 'model',
    name = 'yolov8/detect/val'
)
```

```python
from ultralytics import YOLO
model_path =  '/content/models/yolov8/detect/train3/weights/best.pt'
test_img_dir = '/content/human_detection_dataset/val/images'

model = YOLO(model_path)

for img_name in os.listdir(test_img_dir)[:10]:
  img_path = os.path.join(test_img_dir, img_name)
  model.predict(img_path, save=True)
```
